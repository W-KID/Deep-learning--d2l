```python
import numpy as np
from matplotlib_inline import backend_inline
from d2l import torch as d2l

def f(x):#给出一个公式
    return 3*x**2-4*x


def numerical_lim(f,x,h):#求给定x的fx极限
    return (f(x+h)-f(x))/h
h=0.1 #给定瞬时变化率
for i in range(5):
    print(f'h:={h:.5f},numerical_limit={numerical_lim(f,1,h):.5f}')
    h*=0.1#更新瞬时变化率
# 可以看出当x=1时，导数是2。
```

```python
x = np.arange(0, 3, 0.1)
d2l.plot(x, [f(x), 2 * x - 3], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```
